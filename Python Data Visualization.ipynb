```python
country_data = pd.read_csv('countries.csv')
```

```python
country_data
```

```python
#Compare US and China's population growth
```

```python
us = country_data[country_data.country == 'United States']
```

```python
us
```

```python
china = country_data[country_data.country == 'China']
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(["US", "China"])
plt.title("China vs US Population Growth")
plt.xlabel("Year")
plt.ylabel("Population (In Millions)")
plt.show()
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100 - 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100 - 100)
plt.legend(["US", "China"])
plt.title("China vs US Population Growth")
plt.xlabel("Year")
plt.ylabel("Population Growth %")
plt.show()
```
